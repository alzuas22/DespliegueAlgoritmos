Importamos el conjunto de datos.

```python
import pandas as pd
from sklearn.datasets import fetch_20newsgroups

# Cargamos el conjunto de datos de 20 Newsgroups
newsgroups = fetch_20newsgroups(subset='all')
df = pd.DataFrame({'text': newsgroups.data, 'target': newsgroups.target})
# Mostramos la estructura y contenido del conjunto de datos
print(df.head())
```

Vemos que tiene dos columnas sencillas: el texto y la categoría a la que pertenece el texto. Esta categoría será nuestro target.

Pasamos ahora a Vectorizar el texto

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

# Dividimos los datos en conjuntos de entrenamiento y prueba
train, test = train_test_split(df, test_size=0.2, random_state=123, stratify=df['target'])

# Definimos las características y las etiquetas
x_raw = train['text']
y_raw = train['target']
x_train, x_test, y_train, y_test = train_test_split(x_raw, y_raw, test_size=0.20, random_state=123, stratify=y_raw)

# Creamos un CountVectorizer para vectorizar el texto
vectorizer = CountVectorizer(stop_words='english', max_features=10000)
x_train_vec = vectorizer.fit_transform(x_train)
x_test_vec = vectorizer.transform(x_test)
```

Realizamos varios entrenamientos que van a ser registrados en nuestro entorno de MLflow para luego poder comparar sus métricas.

```python
import mlflow
import mlflow.sklearn
import subprocess
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Lista de valores de n_estimators y min_samples_leaf para iterar
n_estimators = [50, 100, 200, 500]
min_samples_leafs = [2, 4, 10]

# Iniciamos el proceso de la interfaz de usuario de MLflow
mlflow_ui_process = subprocess.Popen(['mlflow', 'ui', '--port', '5000'])

# Establecemos el experimento en MLflow
mlflow.set_experiment('Experimento Random Forest Periodicos')

# Iteramos sobre los diferentes valores de n_estimators y min_samples_leaf
for n in n_estimators:
    for min_samples_leaf in min_samples_leafs:
        with mlflow.start_run(run_name=f'Est: {n}, Leaf: {min_samples_leaf}') as run:
            clf = RandomForestClassifier(n_estimators=n,
                                         min_samples_leaf=min_samples_leaf,
                                         max_depth=10,
                                         max_features='sqrt',
                                         class_weight='balanced',
                                         random_state=123)

            # Creamos un pipeline para la normalización del texto
            preprocessor = Pipeline(steps=[('scaler', StandardScaler(with_mean=False))])

            model = Pipeline(steps=[('preprocessor', preprocessor),
                                    ('classifier', clf)])
            model.fit(x_train_vec, y_train)

            # Predicciones y métricas en el conjunto de prueba
            y_pred_train = model.predict(x_train_vec)
            y_pred_test = model.predict(x_test_vec)

            accuracy_train = accuracy_score(y_train, y_pred_train)
            accuracy_test = accuracy_score(y_test, y_pred_test)
            precision_test = precision_score(y_test, y_pred_test, average='weighted')
            recall_test = recall_score(y_test, y_pred_test, average='weighted')
            f1_test = f1_score(y_test, y_pred_test, average='weighted')

            # Registramos las métricas y parámetros en MLflow
            mlflow.log_param('n_estimators', n)
            mlflow.log_param('min_samples_leaf', min_samples_leaf)
            mlflow.log_param('max_depth', 10)
            mlflow.log_param('max_features', 'sqrt')
            mlflow.log_metric('accuracy_train', accuracy_train)
            mlflow.log_metric('accuracy_test', accuracy_test)
            mlflow.log_metric('precision_test', precision_test)
            mlflow.log_metric('recall_test', recall_test)
            mlflow.log_metric('f1_test', f1_test)
            mlflow.sklearn.log_model(model, 'clf-model')

            print(f'Run with Est: {n}, Leaf: {min_samples_leaf} logged to MLflow')
```

Una vez ya tenemos el modelo funcionando vamos a crear el main.py y el funciones.py

```python
# funciones.py
%%writefile funciones.py
import argparse
import subprocess
import time
import mlflow
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def argumentos():
    parser = argparse.ArgumentParser(description='__main__ de la aplicación con argumentos de entrada.')
    parser.add_argument('--nombre_job', type=str, help='Valor para el parámetro nombre_documento.')
    parser.add_argument('--n_estimators_list', nargs='+', type=int, help='Lista de los valores de n_estimators.')
    parser.add_argument('--min_samples_leafs_list', nargs='+', type=int, help='Lista de los valores de min_samples_leafs.')
    return parser.parse_args()

def load_dataset():
    # Cargamos el conjunto de datos de 20 Newsgroups
    newsgroups = fetch_20newsgroups(subset='all')
    df = pd.DataFrame({'text': newsgroups.data, 'target': newsgroups.target})
    return df

def data_treatment(df):
    # Dividimos los datos en conjuntos de entrenamiento y prueba
    train, test = train_test_split(df, test_size=0.2, random_state=123, stratify=df['target'])

    # Definimos las características y las etiquetas
    x_raw = train['text']
    y_raw = train['target']
    x_train, x_test, y_train, y_test = train_test_split(x_raw, y_raw, test_size=0.20, random_state=123, stratify=y_raw)

    # Creamos un CountVectorizer para vectorizar el texto
    vectorizer = CountVectorizer(stop_words='english', max_features=10000)
    x_train_vec = vectorizer.fit_transform(x_train)
    x_test_vec = vectorizer.transform(x_test)
    return x_train_vec, x_test_vec, y_train, y_test

def mlflow_tracking(nombre_job, x_train_vec, x_test_vec, y_train, y_test, n_estimators, min_samples_leafs):
    # Iniciamos el proceso de la interfaz de usuario de MLflow
    mlflow_ui_process = subprocess.Popen(['mlflow', 'ui', '--port', '5000'])

    # Establecemos el experimento en MLflow
    mlflow.set_experiment(nombre_job)

    # Iteramos sobre los diferentes valores de n_estimators y min_samples_leaf
    for n in n_estimators:
        for min_samples_leaf in min_samples_leafs:
            with mlflow.start_run(run_name=f'Est: {n}, Leaf: {min_samples_leaf}') as run:
                clf = RandomForestClassifier(n_estimators=n,
                                            min_samples_leaf=min_samples_leaf,
                                            max_depth=10,
                                            max_features='sqrt',
                                            class_weight='balanced',
                                            random_state=123)

                # Creamos un pipeline para la normalización del texto
                preprocessor = Pipeline(steps=[('scaler', StandardScaler(with_mean=False))])

                model = Pipeline(steps=[('preprocessor', preprocessor),
                                        ('classifier', clf)])
                model.fit(x_train_vec, y_train)

                # Predicciones y métricas en el conjunto de prueba
                y_pred_train = model.predict(x_train_vec)
                y_pred_test = model.predict(x_test_vec)

                accuracy_train = accuracy_score(y_train, y_pred_train)
                accuracy_test = accuracy_score(y_test, y_pred_test)
                precision_test = precision_score(y_test, y_pred_test, average='weighted')
                recall_test = recall_score(y_test, y_pred_test, average='weighted')
                f1_test = f1_score(y_test, y_pred_test, average='weighted')

                # Registramos las métricas y parámetros en MLflow
                mlflow.log_param('n_estimators', n)
                mlflow.log_param('min_samples_leaf', min_samples_leaf)
                mlflow.log_param('max_depth', 10)
                mlflow.log_param('max_features', 'sqrt')
                mlflow.log_metric('accuracy_train', accuracy_train)
                mlflow.log_metric('accuracy_test', accuracy_test)
                mlflow.log_metric('precision_test', precision_test)
                mlflow.log_metric('recall_test', recall_test)
                mlflow.log_metric('f1_test', f1_test)
                mlflow.sklearn.log_model(model, 'clf-model')

                print(f'Run with Est: {n}, Leaf: {min_samples_leaf} logged to MLflow')
```

```python
# main.py
%%writefile main.py
from funciones import argumentos, load_dataset, data_treatment, mlflow_tracking

def main():
  print("Ejecutamos el main")
  args_values = argumentos()
  df = load_dataset()
  x_train, x_test, y_train, y_test = data_treatment(df)
  mlflow_tracking(args_values.nombre_job, x_train, x_test, y_train, y_test, args_values.n_estimators_list, args_values.min_samples_leafs_list)

if __name__ == "__main__":
  main()
```
